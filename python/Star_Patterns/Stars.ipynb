Simple Star
---

```python
def simple_star(row):
    for i in range(1 , row+1):
        for j in range(1 , row+1):
            if i == j or i+j == row+1:
                print('*', end='')
            else:
                print(' ',end='')
        print()
```

```python
simple_star(10)
```

# Complex Star

```python
import math

def complex_star(n:float = 5.5):
    q = math.tan(math.pi * 0.4)
    w = math.tan(math.pi * 0.2)

    for j in range(math.ceil(n * q), -1, -1):
        for i in range(-math.ceil(0.55 * n * q / w - n), math.ceil(0.55 * n * q / w - n)):
            
            if (
                # Top star arms
                (j <= 0.55 * n * q and j >= (i + n) * w and j >= (n - i) * w)
                
                # Middle star arms
                or (j >= (i + n) * w and j <= (i + n) * q and j <= (n - i) * q)
                
                # Bottom star arms
                or (j <= (n - i) * q and j >= (n - i) * w and j <= (i + n) * q)
            ):
                print("*", end="")
            else:
                print(" ", end="")
        print()
```
