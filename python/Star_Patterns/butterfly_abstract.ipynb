Butterfly
---

```python
def butterfly(row):
    # Upper part
    for i in range(1 , row+1):
        print(
            "*" * i + " " * (2*(row - i)) + "*" * i
        )
    # Lower part
    for i in range(row-1, 0, -1):
        print(
            "*" * i + " " * (2*(row - i)) + "*" * i
        ) 
```

```python
butterfly(6)
```

# Outline rectangle

```python
def rect_outline(row):
    for i in range(row):
        if i == 0 or i == row-1:
            print('*'*row)
        else:
            print( '*' + ' ' * (row-2) + '*')
```

```python
rect_outline(6)
```

# Outline Diamond

```python
def dia_outline(row):
    # Upper part
    for i in range(1, row+1):
        print(' ' * (row-i)+'*'+ ' ' * (2*i-3) + (('*' if i>1 else '')))
    # Lower part
    for i in range(row-1 , 0 , -1):
        print(' ' * (row-i)+'*'+ ' ' * (2*i-3) + (('*' if i>1 else '')))
```

```python
dia_outline(6)
```

# Outline Triangle

```python
def tri_outline(row):
    for i in range(1 , row+1):
        if i == row:
            print('*' *(2*i-1))
        else:
            print(' '*(row-i)+'*'+' '*(2*i-3)+('*' if i>1 else ''))
```

```python
tri_outline(6)
```
