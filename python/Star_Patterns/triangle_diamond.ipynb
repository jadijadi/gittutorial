Triangles
---

```python
def triangle(row):
    for i in range(1 , row+1):
        print(
            " " * (row-i) + "*" * (2*i - 1)
        )
```

```python
triangle(6)
```

```python
def reversed_triangle(row):
    for i in range(row, 0, -1):
        print(
            " " * (row-i) + "*" * (2*i - 1)
        )
```

```python
reversed_triangle(6)
```

# Diamond

```python
def diamond(row):
    # Upper half
    for i in range(1 , row+1):
        print(
            " " * (row-i) + "*" * (2*i - 1)
        )
    # Lower half
    for i in range(row-1, 0, -1):
        print(
            " " * (row-i) + "*" * (2*i - 1)
        )    
```

```python
diamond(6)
```
